```python
import requests
import uuid
import json
import yaml
```

```python
# Step 1: Login with credentials
with open("..\\config\\user_d.json", 'r') as openfile:
    user_d = json.load(openfile)
conn_params =  user_d["conn_params"]
session_id = str(uuid.uuid4())  # Generate unique session ID
login_response = requests.post("http://localhost:8000/login", json={
    "session_id": session_id,
    "conn_params": conn_params
    }
                )
print(login_response.json())
```

```python
with open('..\\config\\First_comp_run.yml', 'r') as openfile:
    playbook_d = yaml.safe_load(openfile)

response = requests.post("http://localhost:8000/run_comparison"
    ,json={"session_id": session_id, 
            "play_compare_d": playbook_d}
    )

response.text
```

```python
# Step 3: Logout when done
requests.post("http://localhost:8000/logout", params={"session_id": session_id})
```
